# Hypothesis 2 검증 예시: 고령층은 미세먼지에 더 민감하게 반응한다.

```python
import pandas as pd
import statsmodels.api as sm
from scipy.stats import pearsonr, spearmanr
from src.utils.db_util import DustDB
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

# ───────────────────────────────────────────────────────────────
#  DB 연결 및 age_group_dim 로드
# ───────────────────────────────────────────────────────────────

db = DustDB()
age_dim = db.load_table("age_group_dim")  # age_group, age_range


# ───────────────────────────────────────────────────────────────
# 천식·비염 fact 테이블 로드 및 age_range 병합
# ───────────────────────────────────────────────────────────────

# pm10_asthma_fact, pm25_asthma_fact
df_pa10 = db.load_table("pm10_asthma_fact").rename(columns={"ym": "date_id"})
df_pa25 = db.load_table("pm25_asthma_fact").rename(columns={"ym": "date_id"})
df_pr10 = db.load_table("pm10_rhinitis_fact").rename(columns={"ym": "date_id"})
df_pr25 = db.load_table("pm25_rhinitis_fact").rename(columns={"ym": "date_id"})

# age_group_dim과 조인하여 age_range 컬럼 추가
df_pa10 = pd.merge(df_pa10, age_dim, on="age_group", how="left")
df_pa25 = pd.merge(df_pa25, age_dim, on="age_group", how="left")
df_pr10 = pd.merge(df_pr10, age_dim, on="age_group", how="left")
df_pr25 = pd.merge(df_pr25, age_dim, on="age_group", how="left")


# ───────────────────────────────────────────────────────────────
#  PM 데이터 로드 (월×지역별 평균 PM10/PM2.5)
# ───────────────────────────────────────────────────────────────

df_pm10 = db.load_table("pm10_fact")[["date_id","region_name","pm10"]].drop_duplicates()
df_pm10 = df_pm10.rename(columns={"region_name": "region"})
df_pm25 = db.load_table("pm25_fact")[["date_id","region_name","pm25"]].drop_duplicates()
df_pm25 = df_pm25.rename(columns={"region_name": "region"})


# ───────────────────────────────────────────────────────────────
# 연령대별 상관·회귀 분석 함수 정의
# ───────────────────────────────────────────────────────────────

def analyze_age_sensitivity(df_disease, pm_df, x_col, label_disease, label_pm):
    """
    df_disease: pm*_<disease>_fact 테이블에 age_range 컬럼이 붙은 DataFrame
    pm_df      : pm10_fact 또는 pm25_fact (date_id, region, pm10/pm25)
    x_col      : "pm10" 또는 "pm25"
    label_disease: "asthma" 또는 "rhinitis"
    label_pm   : "pm10" 또는 "pm25"
    """
    results = []
    for age_label in age_dim["age_range"].unique():
        df_age = df_disease[df_disease["age_range"] == age_label]
        if df_age.empty:
            continue

        monthly_age = (
            df_age
            .groupby(["date_id","region"], as_index=False)[["visit_count"]]
            .sum()
            .rename(columns={"visit_count": "visit_age"})
        )

        df_pm_age = pd.merge(monthly_age, pm_df, on=["date_id","region"], how="inner").dropna(subset=[x_col, "visit_age"])
        if df_pm_age.empty:
            continue

        pear_r, pear_p = pearsonr(df_pm_age[x_col], df_pm_age["visit_age"])
        rho, sp_p = spearmanr(df_pm_age[x_col], df_pm_age["visit_age"])

        X = sm.add_constant(df_pm_age[x_col])
        y = df_pm_age["visit_age"]
        model = sm.OLS(y, X).fit()
        beta = model.params[x_col]
        pval = model.pvalues[x_col]
        r2 = model.rsquared

        results.append({
            "disease":      label_disease,
            "pm_type":      label_pm,
            "age_range":    age_label,
            "pearson_r":    pear_r,
            "pearson_p":    pear_p,
            "spearman_rho": rho,
            "spearman_p":   sp_p,
            "ols_beta":     beta,
            "ols_p":        pval,
            "ols_r2":       r2
        })

    return pd.DataFrame(results)


# ───────────────────────────────────────────────────────────────
# 각 조합별 분석 실행
# ───────────────────────────────────────────────────────────────

# 천식 vs PM10/PM2.5
df_ast_pm10 = analyze_age_sensitivity(df_pa10, df_pm10, "pm10", "asthma", "pm10")
df_ast_pm25 = analyze_age_sensitivity(df_pa25, df_pm25, "pm25", "asthma", "pm25")

# 비염 vs PM10/PM2.5
df_rhn_pm10 = analyze_age_sensitivity(df_pr10, df_pm10, "pm10", "rhinitis", "pm10")
df_rhn_pm25 = analyze_age_sensitivity(df_pr25, df_pm25, "pm25", "rhinitis", "pm25")


# ───────────────────────────────────────────────────────────────
# 결과 통합 및 age_range 순서 지정
# ───────────────────────────────────────────────────────────────

df_hyp2 = pd.concat([df_ast_pm10, df_ast_pm25, df_rhn_pm10, df_rhn_pm25], ignore_index=True)
df_hyp2 = pd.merge(
    df_hyp2,
    age_dim[["age_group","age_range"]],
    on="age_range",
    how="left"
)


# ───────────────────────────────────────────────────────────────
# 결과 출력
# ───────────────────────────────────────────────────────────────

print("=== 가설 2: 연령대별 미세먼지 민감도 결과 ===")
display(df_hyp2)


# ─────────────────────────────────────────────────
# “연령대별 R² 비교” 그래프 (천식 & 비염, PM10 vs PM2.5)
# ─────────────────────────────────────────────────

for disease_label in ["asthma", "rhinitis"]:
    # 해당 질환만 필터한 뒤, age_group 순으로 정렬
    tmp = df_hyp2[df_hyp2["disease"] == disease_label].sort_values("age_group")
    
    plt.figure(figsize=(6,4))
    # PM10 R² 선
    pm10_series = tmp[tmp["pm_type"] == "pm10"]
    plt.plot(
        pm10_series["age_range"],
        pm10_series["ols_r2"],
        marker="o",
        label="PM10"
    )
    # PM2.5 R² 선
    pm25_series = tmp[tmp["pm_type"] == "pm25"]
    plt.plot(
        pm25_series["age_range"],
        pm25_series["ols_r2"],
        marker="s",
        label="PM2.5"
    )
    plt.title(f"{disease_label.capitalize()} vs PM: 연령대별 R² 비교")
    plt.xlabel("연령대")
    plt.ylabel("R²")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()


# ─────────────────────────────────────────────────
# “연령대별 Pearson r 비교” 그래프 (천식 & 비염, PM10 vs PM2.5)
# ─────────────────────────────────────────────────

for disease_label in ["asthma", "rhinitis"]:
    tmp = df_hyp2[df_hyp2["disease"] == disease_label].sort_values("age_group")
    
    plt.figure(figsize=(6,4))
    # PM10 Pearson r 선
    pm10_series = tmp[tmp["pm_type"] == "pm10"]
    plt.plot(
        pm10_series["age_range"],
        pm10_series["pearson_r"],
        marker="o",
        color="blue",
        label="PM10"
    )
    # PM2.5 Pearson r 선
    pm25_series = tmp[tmp["pm_type"] == "pm25"]
    plt.plot(
        pm25_series["age_range"],
        pm25_series["pearson_r"],
        marker="s",
        color="green",
        label="PM2.5"
    )
    plt.title(f"{disease_label.capitalize()} vs PM: 연령대별 Pearson r 비교")
    plt.xlabel("연령대")
    plt.ylabel("Pearson r")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()

plt.tight_layout()
plt.show()
```

```python
# 히트맵 분석 
import seaborn as sns
import matplotlib.pyplot as plt

heatmap_df = df_hyp2.pivot(
    index="age_group",              
    columns=["disease", "pm_type"],
    values="pearson_r"
).sort_index(ascending=True)       

heatmap_df.index = age_dim.set_index("age_group").loc[heatmap_df.index, "age_range"]

heatmap_df.columns = [f"{d}_{p}" for d, p in heatmap_df.columns]

plt.figure(figsize=(8, 6))
sns.heatmap(
    heatmap_df,
    annot=True,
    fmt=".3f",
    cmap="RdBu_r",
    cbar_kws={"label": "Pearson r"},
    linewidths=0.5,
    linecolor="gray"
)
plt.title("연령대별 질환·PM Pearson 상관계수 히트맵")
plt.ylabel("Age Range")
plt.xlabel("Disease_PM Type")
plt.tight_layout()
plt.show()
```

# 해석
두 질환(천식, 비염) 모두 영유아(0–5세) 연령대에서 가장 높은 R²와 Pearson r을 보였고,

**고령층(65세 이상)**에서는 R²도 매우 낮고 Pearson r도 작게 나와서 “미세먼지 단일 요인”만으로는 설명력이 거의 없음을 보여줍니다.

### 천식 vs 비염 차이

천식의 경우, PM10이 PM2.5보다 약간 더 높은 설명력(약간 높은 R², r)을 보이며,

비염은 PM2.5가 PM10보다 더 높은 R², r을 보입니다.

즉, 천식은 PM10에, 비염은 PM2.5에 상대적으로 민감도가 더 높다는 점이 드러납니다.

### 고령층 민감도 가설(가설 2) 결론

“고령층(65세 이상)이 미세먼지에 가장 민감하다”는 원래 가설 2는 기각됩니다.

실제로는 **영유아(0–5세)**와(또는) **중장년층(45–64세)**이 미세먼지 변화에 더욱 민감하게 반응합니다.

특히 ‘45–64세 비염 + PM2.5’ 조합에서 Pearson r ≈ 0.12 및 R² ≈ 0.0146로 가장 높은 값을 보였는데,

이는 “비염 환자 중 45–64세 그룹이 PM2.5 변화에 가장 민감하게 반응한다”는 인사이트를 줍니다.

### 설명력(R²)이 모두 낮다는 의미

R²가 전 연령대에서 0.01~0.03 수준이므로,

“미세먼지가 진료량 변동의 주된 원인”이라고 보기에는 부족합니다.
